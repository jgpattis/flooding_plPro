```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
yrl_10 = np.load('7OFS/10_milli_molar/analysis/lowest_lig_rmsd.npy')
yrl_50 = np.load('7OFS/50_milli_molar/analysis/lowest_lig_rmsd.npy')
yrl_100 = np.load('7OFS/100_milli_molar/analysis/lowest_lig_rmsd.npy')

hba_10 = np.load('7oft/10_mm/analysis/lowest_lig_rmsd.npy')
hba_50 = np.load('7oft/50_mm/analysis/lowest_lig_rmsd.npy')
hba_100 = np.load('7oft/100_mm/analysis/lowest_lig_rmsd.npy')

he9_10 = np.load('7ofu/10_mm/analysis/lowest_lig_rmsd.npy')
he9_50 = np.load('7ofu/50_mm/analysis/lowest_lig_rmsd.npy')
he9_100 = np.load('7ofu/100_mm/analysis/lowest_lig_rmsd.npy')
```

```python
time = np.arange(0, 300.02, 0.01)
```

```python
fig, ax = plt.subplots(3,3, sharex=True, sharey=True, figsize=[12.8, 9.6])
ax[0][0].plot(time, yrl_10 *10, label='yrl 10 mm')
ax[0][1].plot(time, yrl_50 *10, label='yrl 50 mm')
ax[0][2].plot(time, yrl_100*10, label='yrl 100 mm')

ax[1][0].plot(time, hba_10 *10, label='hba 10 mm')
ax[1][1].plot(time, hba_50 *10, label='hba 50 mm')
ax[1][2].plot(time, hba_100 *10, label='hba 100 mm')

ax[2][0].plot(time, he9_10 *10, label='he9 10 mm')
ax[2][1].plot(time, he9_50 *10, label='he9 50 mm')
ax[2][2].plot(time, he9_100 *10, label='he9 100 mm')

ax[2][0].set_xlabel('time (ns)', fontsize=16)
ax[2][1].set_xlabel('time (ns)', fontsize=16)
ax[2][2].set_xlabel('time (ns)', fontsize=16)

ax[0][0].set_ylabel(r'RMSD ($\AA$)', fontsize=16)
ax[1][0].set_ylabel(r'RMSD ($\AA$)', fontsize=16)
ax[2][0].set_ylabel(r'RMSD ($\AA$)', fontsize=16)

ax[0][0].set_title('10 mm', fontsize=20)
ax[0][1].set_title('50 mm', fontsize=20)
ax[0][2].set_title('100 mm', fontsize=20)
fig.savefig('all9_lowest_lig_rmsd.png')
```
